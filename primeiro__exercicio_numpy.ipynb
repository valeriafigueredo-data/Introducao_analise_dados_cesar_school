## 5. Criar funcao Python (def) para criar matriz 2x2 com elemento unico.

Crie uma função chamada de ```cria_matrix()```, onde ela recebe um inteiro.

Ex: ```cria_matrix(2)``` retorna matriz 2x2 com todos elementos 2.

```cria_matrix(12)``` retorna matriz 2x2 com todos elementos 12.

````python
def cria_matrix(n):
  # Matriz 2x2
  xarray = np.full((2, 2), n)
  return xarray
````

````python
# exemplo de saída esperada
m1 = cria_matrix(25)
print(m1)
````

````python
# exemplo 2 de saída esperada
m1 = cria_matrix(12)
print(m1)
````

Faça agora o mesmo processo, mas para elemento geral. Cheque que a entrada do tamanho da matriz NxN seja um número inteiro.

````python
def cria_matrix_geral(n, c1=2, c2=2):
  # coloque seu código ai...

  if type(c1) is int and type (c2) is int:
    narray = np.full((c1, c2), n)
    return narray
  else:
    print ('Informar numeros inteiros')
````

## 7. How to find the memory size of any array

````python
# NumPy array
arr = np.array([[1, 2, 3], [4, 5, 6]])

# Calcula a memória do array em bytes
tamanho_array = arr.nbytes

print(f'Quantos bytes este array oculpa na memória: {tamanho_array:.2f} bytes')

# Conversão
tamanho_memo_kb = tamanho_array / 1024
tamanho_memo_mb = tamanho_memo_kb / 1024
tamanho_memo_gb = tamanho_memo_mb / 1024

print(f'Quantos bytes este array oculpa na memória em megabytes: {tamanho_memo_mb}MB')
print(f'Quantos bytes este array oculpa na memória em kilobytes: {tamanho_memo_kb}KB')
print(f'Quantos bytes este array oculpa na memória em gigabytes: {tamanho_memo_gb}GB')
````

````python
# Ou uma função

def memory_size(arr):
  tamanho_array = arr.nbytes
  return tamanho_array

def converte_bytes(arr):

  tam_bytes = memory_size(arr)
  tamanho_memo_kb =  tam_bytes / 1024
  tamanho_memo_mb = tamanho_memo_kb / 1024
  tamanho_memo_gb = tamanho_memo_mb / 1024
  print(f'Quantos bytes este array oculpa na memória: {tam_bytes} bytes')
  print(f'Quantos bytes este array oculpa na memória em megabytes: {tamanho_memo_mb}MB')
  print(f'Quantos bytes este array oculpa na memória em kilobytes: {tamanho_memo_kb}KB')
  print(f'Quantos bytes este array oculpa na memória em gigabytes: {tamanho_memo_gb}GB')

tamanho_bytes = memory_size(x5)
print(f'Memória em bytes: {tamanho_bytes}\n')
converte_bytes(x5)
````
